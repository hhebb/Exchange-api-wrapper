```python
class Exchange:
    name = ""
    msg = ""
    symbol = ""
    version = ""
    rst = False
    json_time = datetime.datetime.now()
    last_time = datetime.datetime.now()
    def get_json(self, currency_list): # 거래소 별 json 다운로드 로직
        pass
    def get_last(self, currency_list): # 거래소 별 last price parsing 로직
        pass
    def debug(self, error_msg):
        print(error_msg)
    def _run(self, currency_list):
        try:
            self.get_json(currency_list)
            self.json_time = datetime.datetime.now()
        except:
            self.debug("ERROR : %s get_json : %s" % (self.name, self.msg))
            return False
        try:
            self.get_last(currency_list)
            self.last_time = datetime.datetime.now()
            return True
        except Exception as e:
            self.debug("ERROR : %s get_last : %s" % (self.name, self.msg))
            return False
    def run(self, currency_list):
        self._run(currency_list)
    async def async_run(self, currency_list):
        self._run(currency_list)
```

```python
# 거래소별 코인 시세
# 4 개 거래소, 3 개 코인
class Ticker:
    
    # 환율 구하기
    def ex_us_kr(self):
        url = 'https://quotation-api-cdn.dunamu.com/v1/forex/recent?codes=FRX.KRWUSD'
        exchange =requests.get(url).json()
        self.ratio = int(exchange[0]['basePrice'])
    
    def korbit(self):
        dic = {}
        dic['거래소'] = '코빗'
        lst_currency = ['btc','eth','xrp']
        for cur in lst_currency:
            currency = cur +'_krw'
            params = {'currency_pair' : currency}
            response = requests.get('https://api.korbit.co.kr/v1/ticker', params=params)
            res = response.json()
            dic[cur] = {'last':res['last']}
        
        return dic
    
    def polo(self):
        dic = {}
        dic['거래소'] = '폴로닉스'
        lst_currency = ['USDT_BTC','USDT_ETH','USDT_XRP']
        lst_name = ['btc','eth','xrp']
        params = {'command' : 'returnTicker'}
        response = requests.get('https://poloniex.com/public?command=returnTicker', params=params)
        res = response.json()
        for i, cur in enumerate(lst_currency):
            dic[lst_name[i]] = {'last':int(float(res[cur]['last'])*self.ratio)}
            
        return dic
    
    def coinbase(self):
        dic = {}
        dic['거래소'] = '코인베이스'
        lst_currency = ['BTC','ETH','XRP']
        lst_name = ['btc','eth','xrp']
        for i, cur in enumerate(lst_currency):
            params = {'currency' : cur}
            response = requests.get('https://api.coinbase.com/v2/exchange-rates', params=params)
            res = response.json()
            dic[lst_name[i]] = {'last':int(float(res['data']['rates']['KRW']))}
        return dic
    
    def bittrex(self):
        dic = {}
        dic['거래소'] = '비트렉스'
        lst_currency = ['USDT-BTC','USDT-ETH','USDT-XRP']
        lst_name = ['btc','eth','xrp']
        for i, cur in enumerate(lst_currency):
            params = {'market' : cur}
            response = requests.get('https://api.bittrex.com/api/v1.1/public/getticker', params=params)
            res = response.json()
            dic[lst_name[i]] = {'last':int(float(res['result']['Last'])*self.ratio)}
        
        return dic
```

```python
form = uic.loadUiType('C:\\Users\\김민수\\Desktop\\trading_tmp\\crypto_dash_1.ui')[0] # UI 로드

class myWindow(QMainWindow, form):
    def __init__(self):
        super().__init__()
        self.setupUi(self) # UI 셋팅
        self.tick = Ticker()
        self.set_table()
        self.set_timer_collective() # 루프 일괄 실시간 업뎃
        
        # -- UI 이벤트 콜백함수 등록 --
    
    def set_timer_collective(self):
        self.timer_account = QTimer()
        self.timer_account.timeout.connect(self.collecive_update)
        self.timer_account.setSingleShot(False)
        self.timer_account.start(3000)
        
    def set_table(self):
        self.tableWidget_btc.setColumnCount(3)
        self.tableWidget_eth.setColumnCount(3)
        self.tableWidget_xrp.setColumnCount(3)
        self.tableWidget_ltc.setColumnCount(3)
        
        self.tableWidget_btc.setHorizontalHeaderLabels(('거래소','가격','차익률'))
        self.tableWidget_eth.setHorizontalHeaderLabels(('거래소','가격','차익률'))
        self.tableWidget_xrp.setHorizontalHeaderLabels(('거래소','가격','차익률'))
        self.tableWidget_ltc.setHorizontalHeaderLabels(('거래소','가격','차익률'))
    
    
    def collecive_update(self):
        self.tableWidget_btc.setRowCount(0) # 행 갯수 0 으로 초기화
        self.tableWidget_eth.setRowCount(0) # 행 갯수 0 으로 초기화
        self.tableWidget_xrp.setRowCount(0) # 행 갯수 0 으로 초기화
        
        self.tick.ex_us_kr()
        k = self.tick.korbit()
        p = self.tick.polo()
        c = self.tick.coinbase()
        b = self.tick.bittrex()
        
        self.trader = [k,p,c,b]
        
        # btc
        for t in self.trader:
            rowPosition = self.tableWidget_btc.rowCount()
            self.tableWidget_btc.insertRow(rowPosition)
            
            item = QTableWidgetItem(t['거래소'])
            item_price = QTableWidgetItem(str(t['btc']['last']))

            self.tableWidget_btc.setItem(rowPosition, 0, item)
            self.tableWidget_btc.setItem(rowPosition, 1, item_price)
            
        # eth
        for t in self.trader:
            rowPosition = self.tableWidget_eth.rowCount()
            self.tableWidget_eth.insertRow(rowPosition)
            
            item = QTableWidgetItem(t['거래소'])
            item_price = QTableWidgetItem(str(t['eth']['last']))

            self.tableWidget_eth.setItem(rowPosition, 0, item)
            self.tableWidget_eth.setItem(rowPosition, 1, item_price)
        
        # rpx
        for t in self.trader:
            rowPosition = self.tableWidget_xrp.rowCount()
            self.tableWidget_xrp.insertRow(rowPosition)
            
            item = QTableWidgetItem(t['거래소'])
            item_price = QTableWidgetItem(str(t['xrp']['last']))

            self.tableWidget_xrp.setItem(rowPosition, 0, item)
            self.tableWidget_xrp.setItem(rowPosition, 1, item_price)
            
```
